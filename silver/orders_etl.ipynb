```python
# create temporary etl view
from pyspark.sql.functions import col, cast, lit, current_timestamp, year

source_table_abfss_path = "abfss://BUNN_Foundation_NONPROD@onelake.dfs.fabric.microsoft.com/silver_sapecc_lakehouse.Lakehouse/Tables/dbo_1/orders"

source_df = spark.read.format("delta").load(source_table_abfss_path)

temp_view = "orders_etl"

transformed_df = source_df.select(
    col("SalesOrderNumber"),
    col("SalesOrderLineNumber").cast("string").alias("SalesOrderLineNumber"),
    col("OrderDate"),
    col("CustomerName"),
    col("Email"),
    col("Item"),
    col("Quantity"),
    col("UnitPrice").cast("decimal(18,2)").alias("UnitPrice"),
    col("Tax").cast("decimal(18,2)").alias("Tax"),
    lit("I").alias("action_type"),  
    current_timestamp().alias("row_insert_timestamp"), 
    current_timestamp().alias("row_update_timestamp") 
).where(year(col("OrderDate")).isin(2019, 2020))

transformed_df.createOrReplaceTempView(temp_view)

#print(f"NotebookB: Temporary view '{temp_view}' created successfully.")
#mssparkutils.notebook.exit("NotebookB completed and created temp view")    
```
