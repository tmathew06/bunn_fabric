```python
df=spark.read.format("csv").option("header","false").load("Files/inbound-sap/orders/2019.csv")
display(df.head(10))
```

```python
from pyspark.sql.types import *

orderSchema = StructType([
     StructField("SalesOrderNumber", StringType()),
     StructField("SalesOrderLineNumber", IntegerType()),
     StructField("OrderDate", DateType()),
     StructField("CustomerName", StringType()),
     StructField("Email", StringType()),
     StructField("Item", StringType()),
     StructField("Quantity", IntegerType()),
     StructField("UnitPrice", FloatType()),
     StructField("Tax", FloatType())
 ])

df = spark.read.format("csv").schema(orderSchema).load("Files/inbound-sap/orders/*.csv")

display(df.head(10))
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("orders")
```

#### Products

```python
df=spark.read.format("csv").option("header","True").load("Files/inbound-sap/products/*.csv")
display(df.head(10))
```

```python
from pyspark.sql.types import *

productSchema = StructType([
     StructField("ProductID", StringType()),
     StructField("ProductName", StringType()),
     StructField("Category", StringType()),
     StructField("ListPrice", FloatType()),
 ])

df = spark.read.format("csv").schema(productSchema).load("Files/inbound-sap/products/*.csv")

display(df.head(10))
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("products")
display(df.head(10))
```
