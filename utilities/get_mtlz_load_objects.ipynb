```python
# import libraries
import json
```

```python
# define mtls load table path
load_table_path = "abfss://BUNN_Foundation_NONPROD@onelake.dfs.fabric.microsoft.com/utilities_lakehouse.Lakehouse/Tables/materialization_load"
df = spark.read.format("delta").load(load_table_path)

#get active objects from the load table
df_active =df.filter("is_active = True")
active_records = df_active.select("lakehouse_name", "input_source_schema","input_source","output_target_schema","output_target", "load_option").collect()

# convert to JSON
active_records_json = json.dumps([row.asDict() for row in active_records])
#print(active_records_json)

# return the JSON as notebook output
mssparkutils.notebook.exit(active_records_json)
```

```python
spark.stop()
```
